# Forest Fire Prediction – Data Preprocessing

In this notebook, we prepare the Forest Fires dataset for modeling. 
The following preprocessing steps are performed:

1. Remove duplicates
2. Apply log transformation to the target variable
3. Encode categorical variables
4. Scale numerical features
5. Save the cleaned dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
#Load the dataset
df = pd.read_csv("forestfires.csv")
df.head()
```

We check for duplicate rows and remove them to avoid bias in modeling.

```python
print("Before removing duplicates:", df.shape)
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python

print("Duplicate rows:", df.duplicated().sum())


df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)


print("Missing values per column:\n", df.isnull().sum())
```

Duplicate rows are removed to ensure data integrity.
Missing values are checked. No missing values were found.

The target variable `area` is highly right-skewed. We apply a log transformation:
log_area = log(1 + area)

```python
df['log_area'] = np.log1p(df['area'])

df = df.drop(columns=['area'])

df.head()
```

The dataset contains two categorical variables: `month` and `day`.
We use Encoding to convert them to numeric format.

```python
df = pd.get_dummies(df, columns=['month', 'day'], drop_first=True)

df.head()
```

Standardizing numerical features ensures they have mean=0 and std=1.
This is important for algorithms sensitive to feature scales.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_cols = df.select_dtypes(include=np.number).columns

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

df.head()
```

```python
print("Cleaned dataset shape:", df.shape)
print(df.info())
```

```python
df.to_csv("/Users/sa25/Documents/Forest Fires/Forest-Fires/data/forestfires.csv", index=False)
          
print("File saved successfully")
```

# Data Cleaning

In this notebook, the Forest Fires dataset was prepared for modeling by performing essential data cleaning and preprocessing steps.

1. Removed Duplicate Records
   Any duplicate rows were removed to ensure data integrity and prevent bias in the model.

2. Log Transformation of Target Variable
   The target variable `area` is highly right-skewed. To normalize the distribution and reduce the impact of extreme values, a log transformation was applied:  
   `log_area = log(1 + area)`. The original `area` column was then dropped.

3. Encoded Categorical Variables 
   The categorical features `month` and `day` were converted into numeric format

4. Feature Scaling 
 `StandardScaler` was used, to have a mean of 0 and a standard deviation of 1. 

5. Final Check and Save
   The cleaned dataset was inspected for correct data types, absence of missing values, and appropriate shape, then saved.
